```python
import pandas as pd
```

```python
mvps = pd.read_csv('mvps.csv')
```

```python
mvps
```

```python
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
```

```python
mvps.head()
```

```python
players = pd.read_csv("players.csv")
```

```python
del players["Unnamed: 0"]
del players["Rk"]
```

```python
players.head()
```

```python
players["Player"].head(50)
```

```python
players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
players.groupby(["Player", "Year"]).get_group(("Greg Anderson", 1991))
```

```python
def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row
players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combined.head(5)
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined 
```

```python
combined[combined["Player"] == "LeBron James"]
```

```python
# Identify affected rows (e.g., Year >= 2022)
mask = combined["Year"] >= 2022

# Save the misaligned columns
pos_wrong = combined.loc[mask, "Pos"]
age_wrong = combined.loc[mask, "Age"]
tm_wrong = combined.loc[mask, "Tm"]

# The correct data is in the wrong columns, so swap them:
combined.loc[mask, "Pos"] = tm_wrong    # Tm column actually has Pos data
combined.loc[mask, "Age"] = pos_wrong   # Pos column actually has Age data
combined.loc[mask, "Tm"] = age_wrong    # Age column actually has Tm data
```

```python
teams = pd.read_csv("teams.csv")
```

```python
teams
```

```python
# Remove rows where W contains "Division"
teams = teams[~teams["W"].str.contains("Division")]
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False )
```

```python
nicknames = {}

with open("nicknames.txt") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name



```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
stats = combined.merge(teams, how="outer", on=["Team", "Year"])
```

```python
combined[combined["Player"] == "Zach Edey"]
```

```python
stats
```

```python
del stats["Unnamed: 0"]
```

```python
stats = stats.apply(pd.to_numeric, errors="ignore")
```

```python
stats["GB"].unique()
```

```python
stats["GB"] = stats["GB"].str.replace("—", "0")
```

```python
stats["GB"].unique()
```

```python
stats["GB"] = pd.to_numeric(stats["GB"])
```

```python
stats = stats.sort_values(by="Player").reset_index(drop=True)
```
