```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
years = list(range(1991,2026))
```

```python
dfs = []
for year in years:
    with open("advanced_stats/{}.html".format(year), encoding="utf-8") as f:
        page = f.read()
    
    soup = BeautifulSoup(page, 'html.parser')
    soup.find('tr', class_="thead").decompose()
    player_table = soup.find_all(id="advanced")[0]
    player_df = pd.read_html(str(player_table))[0]
    player_df["Year"] = year
    dfs.append(player_df)
```

```python
player_advanced = pd.concat(dfs)
```

```python
player_advanced.to_csv("player_advanced_og.csv", index=False)
```

```python
def single_row(df):
    if df.shape[0] == 1: 
        return df
    else:
        row = df[df["Tm"].isin(["2TM", "3TM", "4TM", "5TM", "6TM", "7TM", "8TM", "9TM", "10TM"])]
        return row
player_advanced = player_advanced.groupby(["Player", "Year"]).apply(single_row)
```
